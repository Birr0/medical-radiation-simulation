```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

plt.style.use("science")
```

```python
phantom_length = 40 # cm
unit_length = phantom_length/1000 # cm
```

```python
doses = np.array([np.genfromtxt(f'./outputs/job{i}.csv', delimiter=',')[:, -1] for i in range(1,11)])
bragg_peaks = np.array([
    unit_length * np.argmax(doses, axis=1)
])
np.mean(bragg_peaks), np.std(bragg_peaks)
```

```python
plt.figure(figsize=(8,6))

plt.errorbar(
    [unit_length*i for i in range(1000)],
    np.mean(doses, axis=0),
    yerr=np.std(doses, axis=0),
    label="Dose"
)

plt.xlabel("Depth / cm", fontsize=18)
plt.ylabel("Dose / Gy", fontsize=18)
plt.legend()

plt.savefig("./plots/Dose_Depth_ex3_1.png", dpi=300)
```
