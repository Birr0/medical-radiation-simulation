```python
import subprocess
import os
import numpy as np 
import scienceplots
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use("science")

EXEC_PATH = "/Applications/topas/bin/topas "

os.environ["TOPAS_G4_DATA_DIR"] = "/Applications/G4Data"
os.environ["QT_QPA_PLATFORM_PLUGIN_PATH"] = "/Applications/topas/Frameworks"

def topas(filepath):
    return subprocess.run(EXEC_PATH + filepath, env=os.environ, shell=True)

def beam_energy_file(original_fp, new_fp, beam_energy, output_fp):
    try:
        with open(original_fp, 'r') as original:
            with open(new_fp, 'w') as new:
                new.write(original.read())
                new.write(f'\n\ni:Ts/Seed = 0\nd:So/Demo/BeamEnergy = {beam_energy} MeV\ns:Sc/DoseScorer/OutputFile = "{output_fp}"')
    except FileNotFoundError:
        print("File not found.")
    return new_fp

# Example usage:
```

```python
phantom_length = 40 # cm
delta = 1000
unit_length = phantom_length/delta # cm
```

```python
b_peaks = []
b_peak_unc = []
unit_length = 40/100

for beam_energy in np.arange(3000, 6000, 500):
    doses = []
    
    topas(
        beam_energy_file("./base.txt", f"./jobs/{beam_energy}.txt", beam_energy=beam_energy, output_fp=f'./outputs/job{beam_energy}')
    )
    dose = np.genfromtxt(f'./outputs/job{beam_energy}.csv', delimiter=',')[:, -1]
    b_peaks.append(40 - unit_length * np.argmax(dose))

    #b_peak_unc.append(unit_length * np.std(np.argmax(doses, axis=1)))
    #print(40 - unit_length * np.argmax(dose))
    plt.scatter(
        [40 - unit_length * i for i in range(0,100)],
        dose
    )

    plt.show()
```

```python
# Bragg Peak with no inset present (30.4, 0.07589466384404117)
```

```python
f = lambda x, m, c: m*x+c
x = np.arange(3000, 6000, 500)

m, c = curve_fit(f, x, b_peaks)[0]
plt.figure(figsize=(8,6))

plt.scatter(
    x,
    b_peaks
)
plt.plot(
    x,
    m*x+c,
    c="red",
    ls="--",
    label=f"{round(m,3)} BE {round(c,3)}"
)

plt.axhline(y = 30.4, label="Bragg Peak")
plt.fill_between(x, 30.4 - 0.08, 30.4 + 0.08, color='lightblue', alpha=0.5, label="BP uncertainty")

plt.tick_params(axis='both', which='major', labelsize=12)
plt.xlabel("Beam energy / MeV", fontsize=18)
plt.ylabel("Bragg Peak position / cm", fontsize=18)

plt.xlim(2975, 5525)
d = (30.4 - c)/m
max_ = (30.48 - c)/m - d
print(f"The beam energy is: {d} +- {max_} MeV")
plt.legend(loc="lower right")
plt.savefig("./plots/bp_be.png", dpi=300)
```
